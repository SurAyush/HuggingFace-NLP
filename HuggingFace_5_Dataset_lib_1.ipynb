```python
# rename column
drug_dataset = drug_dataset.rename_column(original_column_name='Unnamed: 0', new_column_name='id')
drug_dataset
```

```python
# let us find out the number of unique drugs and unique conditions
len(drug_dataset['train'].unique('condition'))
```

```python
len(drug_dataset['train'].unique('drugName'))
```

```python
# let us normalize the condition using lowercase
def tolowercase(row):
  return {"condition": row['condition'].lower()}                 # returns a single key value pair that overrides in dictionary

drug_dataset = drug_dataset.map(tolowercase)                # if batched is not passed, it passes one example at a time
```

```python
# let us remove all rows with None as condition
# filter function
drug_dataset = drug_dataset.filter(lambda x: (x['condition'] is not None and x['drugName'] is not None))
drug_dataset
```

```python
# applying normalization now
def tolowercase(row):
  return {"condition": row['condition'].lower()}                 # returns a singe key value pair that overrides in dictionary

drug_dataset = drug_dataset.map(tolowercase)
```

this takes some time, as .map() has batched as false one row is passed at a time
#### list comprehension is much faster than for loops

```python
def tolowercase2(rows):
    # see the sample[:3], the dictionary contains keys as column names and values as list of 3 value
    # so we process and return in that format only
    return {"drugName":[drugname.lower() for drugname in rows['drugName']]}


drug_dataset = drug_dataset.map(tolowercase2, batched = True, num_proc = 8)
```


#### num_proc is very useful when dealing with functions that does not perform parallel processing on it's own
#### batched: True, has default batch size as 1000

we reduced time from 31.4s to 2.38s

```python
# let's create a new column for review size
drug_dataset = drug_dataset.map(lambda x: {"review_length": len(x["review"].split())})
drug_dataset
```

```python
drug_dataset['train'].sort('review_length')[:5]          # sorting by review length
```

```python
# these one word review is not informative enough to predict the condtion
# slice the dataset
drug_dataset = drug_dataset.filter(lambda x: x['review_length']>30)
drug_dataset
```

```python
drug_dataset
```

```python
# lastly let's process html codes
import html
def removehtmlseq(rows):
  return {'review':[html.unescape(rev) for rev in rows['review']]}

drug_dataset = drug_dataset.map(removehtmlseq, batched=True)
```

#### Error explanation

The problem is that we’re trying to mix two different datasets of different sizes: the drug_dataset columns will have a certain number of examples, but the tokenized_dataset we are building will have more (because we are tokenizing long reviews into more than one example by using return_overflowing_tokens=True). That doesn’t work for a Dataset, so we need to either remove the columns from the old dataset or make them the same size as they are in the new dataset. We can do the former with the remove_columns argument:
<br><br>
In the format of Dataset class for say 5 examples <br>
Input: "review" : [] list of 5 examples  <br>
Output: "input_ids" : [] list of say 10 examples

```python
# option 1: remove previous columns
tokenized_dataset = drug_dataset.map(
    tokenize_and_split, batched=True, remove_columns=drug_dataset["train"].column_names
)
```

```python
len(tokenized_dataset["train"]), len(drug_dataset["train"])
```

```python
# alternative: making the old columns the same size as the new ones
def tokenize_and_split(examples):
    result =  tokenizer(
        examples["review"],
        truncation=True,
        max_length=128,
        return_overflowing_tokens=True,
    )

    sample_map = result.pop("overflow_to_sample_mapping")
    # sample map = [0,0,1,2,2,2]
    # It signifies example[0] has split into 2 parts, example[1] has split into 1 and example[2] has split into 3 parts
    for key, values in examples.items():
        result[key] = [values[i] for i in sample_map]
    return result
```

# Apache Arrows >>>

It is a special format to store the data. It stores data in columnar form which helps in faster computation for data analytics works (sum of elements of a column)

## Key Features:

- Columnar Data Storage
- Zero-Copy Conversions: Using *Memory Mapped Structure* amd *Standardized In-Memory Format*
- Memory Mapped Structure: With memory-mapped structures, Apache Arrow does not copy data into RAM. Instead, it creates a "window" (a direct link) to the file stored on disk, allowing it to read data only when needed. That's why Datasets.map() do not melt your RAM even if dataset is too large
- Standardized In-Memory Format: Apache Arrow stores data in a common, columnar format that all tools can use without conversion. Zero-copy sharing between Pandas, NumPy, PyTorch, and Spark. Data is already structured optimally → No need for transformation.
- Cross-Language Compatibility: Arrow’s standard memory format allows seamless data sharing between multiple languages without extra conversion. Traditional data formats (e.g., CSV, JSON) require conversion when used across languages (Python ↔ Java ↔ R)
- Vectorized Processing (SIMD Acceleration): Apache Arrow takes advantage of modern CPU features like SIMD (Single Instruction, Multiple Data) to process entire columns at once instead of one value at a time.
- Compression: Arrow supports modern compression algorithms like LZ4, ZSTD, and Snappy
- IPC (Inter Process Communication): Arrow allows different processes (Python ↔ C++ ↔ Java) to exchange data in memory without copying.
- Parallel Procssing: Apache Arrow supports multi-threaded & parallel processing for large datasets.

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name, token="<TOKEN>" )
model = AutoModelForSeq2SeqLM.from_pretrained(model_name, token="<TOKEN>")
```

```python
drug_dataset_clean.set_format("pandas")
```

```python
drug_dataset_clean
```

```python
df_pd = drug_dataset_clean["train"][:1000]
df_pd
```

```python
def summarize(text, max_length=200, min_length=50):
    inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs.input_ids, max_length=max_length, min_length=min_length, length_penalty=2.0, num_beams=4)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
grouped_reviews = df_pd.groupby("drugName")["review"].apply(lambda x: " ".join(x)).reset_index()
grouped_reviews
```

```python
grouped_reviews["summary"] = grouped_reviews['review'].apply(lambda x: summarize(x))
```

```python
from datasets import Dataset

summary_dataset = Dataset.from_pandas(grouped_reviews)             # only first 100 examples of train split is summarized -- to save time
```
